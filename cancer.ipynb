```python
# Import everything you need to start the project
import pandas as reader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf

# Check if importing is properly completed
print("import complete")
```

```python
# Read the tumor data and store it in table
table = reader.read_csv("/content/tumour/data.csv")

# Check if there is no problem with table
display(table)
```

```python
# That id column and Unnamed: 32 column is unnecessary for our project
# Let's drop them out of the table
table = table.drop(["id", "Unnamed: 32"], axis = 1)

# Check the table again to see if it is properly formatted
display(table)
```

```python
# "M" and "B" alone are not in usable format by the code
# So we change them into M = 1 and B = 0
table["diagnosis"] = table["diagnosis"].map({"M" : 1, "B" : 0})

# Let's check if we changed them correctly
display(table)
```

```python
# So basically the independent variables are everything but the "diagnosis" column
# which determines the outcome of the "diagnosis"
x = table.drop("diagnosis", axis = 1)

# And dependent variable is the result of the "diagnosis", which is either the 0 or 1
y = table["diagnosis"]
```

```python
# Over here, we randomly select 80% of the rows to train our AI model
# And we will use the remaining 20% of the rows to test our AI model
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)

print(x_train.shape)
```

```python
# Some of the variables are much larger on scale than the others and vise versa
# Example: area_mean vs. smoothness_mean
# This will make one variable more impactful than the other in the diagnosis
# So we even them out on more equal impact on diagnosis
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# This is the AI model!
# First layer (input layer): we define the shape of the input for the model to the number of the features in x_train (don't forget the comma)
# Second layer: basic shallow network of 30 neurons (example: when radius is high and area is large, then it is likely to be malignant)
# Third layer: more in depth network of 15 neurons (example: ...but only when symmetry is low too. Then it is even more likely to be malignant)
# Fourth layer (output layer): 1 neuron, makes decision based on all the relationships it gets (this is malignent, this is benign)
# relu: rectified linear unit - turns negative signal into 0
# sigmoid: squashes the output into between the range of 0 and 1
model = tf.keras.Sequential([tf.keras.Input(shape=(x_train.shape[1],)),
                             tf.keras.layers.Dense(30, activation = "relu"),
                             tf.keras.layers.Dense(15, activation = "relu"),
                             tf.keras.layers.Dense(1, activation = "sigmoid")])
```

```python
# instruction on HOW to train the model
# adam adjusts the model's weight to improve prediction
# binary crossentropy: loss function. The model tries to minimize this number
# Accuracy is the metric we want to measure (between 0 and 1)
model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
# We now train the model 100 times
model.fit(x_train_scaled, y_train, epochs = 100, verbose = 1)
```

```python
# Now we use the actual test set to predict probabilities of our trained model
y_pred_probs = model.predict(x_test_scaled)

y_pred = []

# If the probability is greater than 0.5, we set it to 1 (malignant)
# Otherwise, set it to 0
for prob in y_pred_probs:
  if prob > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)

# Print the final accuracy of the model
print("accuracy: ", accuracy_score(y_test, y_pred))
```
